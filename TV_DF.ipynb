```python
# Import the CSV File into the DataFrame
import pandas as pd
```

```python
national = pd.read_csv ('Resources/National Ratings Week of September 14 2020.csv')  
national = pd.DataFrame(national)
national = national.rename(columns={'RANK': 'rank', 'PROGRAM': 'program', 'NETWORK': 'network', 'RATING': 'rating', 'VIEWERS (000)': 'viewers_thou'})
sample = 'National'
national['sample']= sample
national['viewers_thou'] = national['viewers_thou'].replace(',','', regex=True)
national
```

```python
hispanic = pd.read_csv ('Resources/Hispanic Ratings Week of September 14 2020.csv')  
hispanic = pd.DataFrame(hispanic)
hispanic = hispanic.rename(columns={'RANK': 'rank', 'PROGRAM': 'program', 'NETWORK': 'network', 'RATING': 'rating', 'VIEWERS (000)': 'viewers_thou'})
sample = 'Hispanic'
hispanic['sample']= sample
hispanic['viewers_thou'] = hispanic['viewers_thou'].replace(',','', regex=True)
hispanic
```

```python
black = pd.read_csv ('Resources/Black Ratings Week of September 14 2020.csv')  
black = pd.DataFrame(black)
black = black.rename(columns={'RANK': 'rank', 'PROGRAM': 'program', 'NETWORK': 'network', 'RATING': 'rating', 'VIEWERS (000)': 'viewers_thou'})
sample = 'Black'
black['sample']= sample
black['viewers_thou'] = black['viewers_thou'].replace(',','', regex=True)
black
```

```python
nat_hisp = national.append(hispanic, ignore_index=True)
all_ratings = nat_hisp.append(black, ignore_index=True)
all_ratings
```

```python
all_ratings.to_csv('Resources/combined_tv_ratings.csv')
```
