```python
import pandas as pd
import numpy as np
from pprint import pprint
from sqlalchemy import create_engine
from keys import sqlkey
```

```python
#READ IN CSV COLUMNS FOR FIRST FILE

toptv = pd.read_csv ('Resources/combined_tv_ratings.csv') 
ratings_df = pd.DataFrame(toptv, columns= ['rank', 'program','network', 'rating', 'viewers_thou','sample'])
ratings_df
```

```python
#READ IN 2nd CSV

census = pd.read_csv ('Resources/US Census 2017 ACS.csv') 
census_df = pd.DataFrame(census)
census_df

```

```python
#ADD 3 COLUMNS AND CHANGE COLUMN HEADERS TO LOWER CASE
census_df= census_df.rename(columns=str.lower)
census_df['blackdemo'] = 'Black'
census_df['hispanicdemo'] = 'Hispanic'
census_df['nationaldemo'] = 'National'
```

```python
#CREATE DATABASE CONNECTION
engine = create_engine('postgresql://postgres:'+sqlkey+'@localhost:5432/ETL')
connection = engine.connect()
```

```python
#PUSH DATAFRAME TO POSTGRESQL
ratings_df.to_sql(name='toptv', con=connection, if_exists='append', index=False)
```

```python
#PUSH DATAFRAME TO POSTGRESQL
census_df.to_sql(name='census', con=connection, if_exists='append', index=False)
```
